```python
import pandas as pd
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({
    'Person': people,
    'Age': age,
    'Height': height,
    'Weight': weight
})
print(healthcare_data)
```

**SCATTER PLOT**

The scatter plot visually represents the relationship between age and weight/height. Each point on the plot corresponds to a person's age along with their weight and height. The plot helps observe if there is any correlation or pattern between these variables.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.scatter(healthcare_data['Age'], healthcare_data['Weight'], c='blue', label='Weight')
plt.scatter(healthcare_data['Age'], healthcare_data['Height'], c='green', label='Height')

plt.xlabel('Age')
plt.ylabel('Weight / Height')
plt.title('Age vs Weight/Height')
plt.legend()
plt.show()
```

**BAR CHART**

The bar chart compares the weight and height of each person in the dataset. It provides a clear visual representation of how weight and height vary across different individuals.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
bar_width = 0.3
ages = range(len(healthcare_data['Person']))

plt.bar(ages, healthcare_data['Weight'], width=bar_width, color='blue', label='Weight')
plt.bar([pos + bar_width for pos in ages], healthcare_data['Height'], width=bar_width, color='green', label='Height')

plt.xlabel('Person')
plt.ylabel('Weight / Height')
plt.title('Weight and Height by Person')
plt.xticks([pos + bar_width / 2 for pos in ages], healthcare_data['Person'])
plt.legend()

plt.show()
```

**Histogram**

The histogram shows the distribution of ages in the healthcare dataset. It helps visualize the frequency or count of individuals within different age ranges, providing insights into the age distribution of the dataset.*italicized text*

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(healthcare_data['Age'], bins=10, color='skyblue', edgecolor='black')

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Ages in Healthcare Dataset')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(height, bins=20, color='blue', alpha=0.7, label='Height')
plt.hist(weight, bins=20, color='red', alpha=0.7, label='Weight')
plt.title('Histogram of Height and Weight')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
